```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from datetime import date, time, datetime
```

```python
train_users = pd.read_csv("train_users.csv")
```

```python
train_users.head()
```

```python
train_users.describe()
```

```python
dates_list = []
year_list = []
month_list = []

for i in train_users["date_account_created"]:
    date_time_obj = datetime.strptime(i, '%Y-%m-%d')
    year = date_time_obj.date().year
    month = date_time_obj.date().month
    dates_list.append(date_time_obj)
    year_list.append(year)
    month_list.append(month)

train_users["datetime_account_created"] = dates_list
train_users["year_account_created"] = year_list
train_users["month_account_created"] = month_list
```

```python
dates_list = []
year_list = []
month_list = []

for i in train_users["date_first_booking"]:
    if isinstance(i, float):
        date_time_obj = np.nan
        year = np.nan
        month = np.nan
    else:
        date_time_obj = datetime.strptime(i, '%Y-%m-%d')
        year = date_time_obj.date().year
        month = date_time_obj.date().month
    dates_list.append(date_time_obj)
    year_list.append(year)
    month_list.append(month)
    
train_users["datetime_first_booking"] = dates_list
train_users["year_first_booking"] = year_list
train_users["month_first_booking"] = month_list
```

```python
dates_list = []
year_list = []
month_list = []

for i in train_users["timestamp_first_active"]:
    time_stamp_obj = pd.to_datetime(i , format='%Y%m%d%H%M%S')
    year = time_stamp_obj.date().year
    month = time_stamp_obj.date().month
    dates_list.append(time_stamp_obj)
    year_list.append(year)
    month_list.append(month)
    
train_users["datetime_first_active"] = dates_list
train_users["year_first_active"] = year_list
train_users["month_first_active"] = month_list
```

```python
train_users["booked_after_active"] = train_users["datetime_first_booking"] - train_users["datetime_first_active"]
train_users["booked_after_created"] = train_users["datetime_first_booking"] - train_users["datetime_account_created"]
train_users["active_after_created"] = train_users["datetime_first_active"] - train_users["datetime_account_created"]
```
